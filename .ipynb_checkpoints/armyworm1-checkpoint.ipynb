# Background

The concept of integrated pest management (IPM), a sustainable strategy for managing pests, has been in practice for a long time. Pest control and management is a multifacted activity, although it comprises of many factors from insect genetics and pesticide resistance to chemical formulation of pesticides, controlling pest by determining the relationship of their evolutionary biology and its dependence on ecological variables, like temperature, has been by far the most most convenient way of handling an outbreak. It has been noted that the problem of agricultural damage from insects only occurs within a short period of time.

This project is focused on 
Insects are pests and our customers try hard to keep the bugs from ruining their livelihoods. But even the most damaging of insects is only a problem for a short window of time. In the case of Fall Armyworm (Spodoptera frugiperda) - one of the most destructive global species - the egg, pupae, and adult cause no direct damage to crops. Killing these stages is satisfying, but largely inefficient either because they are inert (egg/pupae) or able to flee treated areas (adults). 
Only the larval stage causes damage and is in a position in which treatment is truly effective. But growers should be careful: Spray too early and many of the eggs laid will not have hatched. Spray too late and the larvae will have reached their most destructive period: a time in which much of a crops value is destroyed and the insects are most resistant to chemical treatments.
A tool that helps gauge when a population is in an optimal state for treatment would be helpful for growers. Current degree day models provide broad, deterministic evaluations of developmental trends in a pest population, but do not account for variable responses to temperatures within the population. We propose a new approach that accounts for varying development rates and estimates the proportion of a population in a given lifestage. The hope is to empower growers with knowledge of the likely state of pest populations in their area and estimate when chemical applications are most effective.

# Determininstic Approach

## First order kinetics from eggs to instar 1 (relative to number of eggs remaining)

Model assumptions:
The first order kinetics is one of the simplest models, however, it is too simple to be useful in most circumstances. As such, it makes the following assumptions: 
1. the population is homogeneous (i.e., all members are identical).
2. the population inhabits a uniform and unvarying environment.
3. there are no spatial limitations, and changes are density-independent.

Note: Realistically, population growth is limited by various factors from resource availability to predation. Additional limitations arise from the systemâ€™s internal dynamics, such as overcrowding. First order kinetics may be able to describe the growth of a population for a limited period of time; however, unrestrained growth is never sustainable, and thus additional components are necessary to obtain a more realistic model.

The 1st order kinetics is described as:
\begin{equation*}
\frac{dn}{dt} = -(k_1)(n)
\end{equation*}

development rate constant for instar 1:
\begin{equation*}
k_1
\end{equation*}

## Solution to first order kinetics

\begin{equation*}
ln(\frac{n_1}{n_0}) = -(k_1)(t-t_0)
\end{equation*}

Number of eggs at time = 0:
\begin{equation*}
n_0
\end{equation*}

Number of remaining eggs at time = 1:
\begin{equation*}
n_1
\end{equation*}

based on the paper the development rate for 1st instar is:
\begin{equation*}
k_1 = 0.0213T - 0.1817
\end{equation*}

so:
\begin{equation*}
ln(\frac{n_{eggs_{final}}}{n_{eggs_{initial}}})= -(0.0213T - 0.1817)(t-t_0)
\end{equation*}

the proportion of eggs:
\begin{equation*}
\frac{n_{eggs_{final}}}{n_{eggs_{initial}}} = e^{-(0.0213T - 0.1817)(t-t_0)}
\end{equation*}

the proportion of 1st instar:
\begin{equation*}
\frac{n_{instar1_{final}}}{n_{instar1_{initial}}} = e^{-(0.0323T - 0.359)(t-t_0)}
\end{equation*}

the proportion of 2nd instar:
\begin{equation*}
\frac{n_{instar2_{final}}}{n_{instar2_{initial}}}= e^{-(0.0465T- 0.6299)(t-t_0)}
\end{equation*}

the proportion of 3rd instar:
\begin{equation*}
\frac{n_{instar3_{final}}}{n_{instar3_{initial}}}= e^{-(0.0404T - 0.0532)(t-t_0)}
\end{equation*}

the proportion of 4th instar:
\begin{equation*}
\frac{n_{instar4_{final}}}{n_{instar4_{initial}}} = e^{-(0.0273T - 0.2996)(t-t_0)}
\end{equation*}

the proportion of 5th instar:
\begin{equation*}
\frac{n_{instar5_{final}}}{n_{instar5_{initial}}} = e^{-(0.0312T - 0.4682)(t-t_0)}
\end{equation*}

the proportion of 6th instar:
\begin{equation*}
\frac{n_{instar6_{final}}}{n_{instar6_{initial}}} = 1- e^{-(0.0312T - 0.4682)(t-t_0)}
\end{equation*}

```python
import math
import datetime

def proportion_eggs(temp, start_date, end_date):
    start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
    end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
    delta =  (end_date-start_date).days
    dev = 0.0213*temp - 0.1817
    prop = math.exp(-1*dev*int(delta))
    return prop
        
def proportion_instar_1(temp, start_date, end_date):
    start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
    end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
    delta =  (end_date-start_date).days
    dev = 0.0323*temp - 0.359
    prop = math.exp(-1*dev*int(delta))
    return prop

def proportion_instar_2(temp, start_date, end_date):
    start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
    end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
    delta =  (end_date-start_date).days
    dev = 0.0465*temp - 0.6299
    prop = math.exp(-1*dev*int(delta))
    return prop
        
def proportion_instar_3(temp, start_date, end_date):
    start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
    end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
    delta =  (end_date-start_date).days
    dev = 0.0404*temp - 0.0532
    prop = math.exp(-1*dev*int(delta))
    return prop
        
def proportion_instar_4(temp, start_date, end_date):
    start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
    end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
    delta =  (end_date-start_date).days
    dev = 0.0273*temp - 0.2996
    prop = math.exp(-1*dev*int(delta))
    return prop
        
def proportion_instar_5(temp, start_date, end_date):
    start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
    end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
    delta =  (end_date-start_date).days
    dev = 0.0312*temp - 0.4682
    prop = math.exp(-1*dev*int(delta))
    return prop
        
def proportion_instar_6(temp, start_date, end_date):
    start_date = datetime.datetime.strptime(start_date, "%Y-%m-%d").date()
    end_date = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
    delta =  (end_date-start_date).days
    dev = 0.0312*temp - 0.4682
    prop = math.exp(-1*dev*int(delta))
    return 1-prop
```

```python
proportion_eggs(30, start_date='2008-08-18', end_date='2008-9-01')
```

```python
start_date = ['2008-08-18','2008-08-20','2008-08-22','2008-08-24','2008-08-26','2008-08-28']
temp = [25,27,29,33,35]
isotherm_plot = []
for i in temp:
    for j in start_date:
        proportion = proportion_eggs(i, start_date=j, end_date='2008-9-01')
        isotherm_plot.append([i,datetime.datetime.strptime(j, "%Y-%m-%d").date(),datetime.datetime.strptime('2008-9-01', "%Y-%m-%d").date(),proportion])
```

```python
import pandas as pd
isotherm = pd.DataFrame(isotherm_plot, columns  = ['temp','start_date','end_date','proportion'])
isotherm['duration'] = isotherm.end_date-isotherm.start_date
isotherm_plot = isotherm[['temp','duration','proportion']]
isotherm_plot
```

```python
import seaborn as sns
ax = sns.lineplot(data=isotherm_plot, x="duration", y="proportion", hue="temp", palette = 'tab10')
ax.set_title('Isotherm Plots for Proportion of Eggs as Function of Time')
```
